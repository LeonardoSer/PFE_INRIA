```python
import itertools
import random
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path 
import sys
import threading
import os
from IPython.display import clear_output
```

```python
x,y = ['a', 'b', 'c', 'd', 'e', 'f', 'g'],[1, 20, 2, 12, 26, 44, 45]
plt.figure(figsize=(8, 5), dpi=80)
plt.xlabel("year")
plt.ylabel("#new_collaborations/#new_authors")


tmp_dist = []
for i in range(len(y)-1):
    tmp_dist.append(abs(y[i] - y[i+1]))
    
stretch_dist = [y[0], tmp_dist [0]]
for i in range(1, len(tmp_dist)):
    stretch_dist.append(tmp_dist[i] + stretch_dist[i])    
stretch_dist

plt.scatter(stretch_dist,y)
plt.xticks(stretch_dist, x)
plt.show
```
