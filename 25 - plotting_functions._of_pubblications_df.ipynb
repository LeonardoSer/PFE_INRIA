```python
import itertools
import random
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path 
import sys
import threading
import os
from IPython.display import clear_output
```

```python
## Computer science authors in Nice
collaboration_df = pd.read_csv('myDATA/00-collaboration_df.csv')
collaboration_df
```

```python
publications_df = pd.read_csv('myDATA/02-publication_df.csv')
publications_df
```

```python
#find the number of new authors in the publication dataset for each year
YEARS = [str(year) for year in range(1990,2019)]
df = publications_df
plt.figure(figsize=(8, 6), dpi=80)
plt.xlabel("year")
plt.ylabel("num_new_authors")
plt.title("num of new authors for each year in publication dataset\n", fontdict = {'fontsize' : 20})
x,y = [], []
for year in YEARS:    
    n_not_new = len(df)
    df = df.loc[df[year] == 0]
    x.append(year)
    y.append(n_not_new - len(df))
plt.xticks(rotation='vertical')
plt.plot(x, y)
plt.scatter(x,y)
plt.savefig('myDATA/num_new_auths_by_year_in_pubblicationDF.png')
plt.show()
```

```python
#find the number of new pubblication in the publication dataset for each year
YEARS = [str(year) for year in range(1990,2019)]
df = publications_df
plt.figure(figsize=(8, 6), dpi=80)
plt.xlabel("year")
plt.ylabel("num_new_publications")
plt.title("num of new publications for each year in publication dataset\n", fontdict = {'fontsize' : 20})

x,y = [], []
for year in YEARS:    
    x.append(year)
    y.append(df[year].sum())
    
plt.xticks(rotation='vertical')
plt.plot(x, y)
plt.scatter(x,y)
plt.savefig('myDATA/num_pubblications_by_year_in_pubblicationDF.png')
plt.show()
```
