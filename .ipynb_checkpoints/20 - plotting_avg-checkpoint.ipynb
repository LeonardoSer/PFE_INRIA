```python
import itertools
import random
import json
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path 
import sys
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
#Symply plot degree trajectories contained in a given dataframe
def plot_avg_df(df, year, kind_of_authors):
    legend = []
    plt.figure(figsize=(8, 6), dpi=80)
    plt.xlabel("year")
    plt.ylabel("num_Collaborations")
    plt.title("# vertex trajectories for " + kind_of_authors + " authors who started in: "+ year + "\n", fontdict = {'fontsize' : 20})
    x=[]
    y=[]
    for i in df:
        if(i!="ID"):
            y.append(df[i].mean())
            x.append(i)

    plt.xticks(rotation='vertical')
    if(len(y)==1):
        plt.scatter(x,y)
    else:
        plt.plot(x, y)
    plt.show()
```

```python
## Read all "10-collaborations_by_year_for_active" datasets and plot them
YEARS = [str(year) for year in range(1990,2019)]

for y in YEARS:
    df_y = pd.read_csv('myDATA/10-collaborations_by_year_for_active/collabs_by_year_'+y+'.csv')
    plot_avg_df(df_y, y, "active")
```

```python
## Read all "10-collaborations_by_year_for_active" datasets and plot them
YEARS = [str(year) for year in range(1990,2019)]

for y in YEARS:
    df_y = pd.read_csv('myDATA/12-collaborations_by_year_for_most_active/collabs_by_year_'+y+'.csv')
    plot_avg_df(df_y, y, "very active")
```
