```python
YEARS = [str(year) for year in range(1990,2019)]

## Computer science authors in France
collaboration_df = pd.read_csv('myDATA/00-collaboration_df.csv')
publications_df = pd.read_csv('myDATA/02-publication_df.csv')
```

```python
## Find the starting collaboration year for each author
def get_ending_Y(aut):
    YEARS = [str(year) for year in range(1990,2019)]
    for y in YEARS[::-1]:
        val = aut[y].values[0]
        if(val != 0):
            return y
        
# add starting collaboration year to each row of the dataset
def add_ending_y(df, pub_df):
    df["ending_year"] = [0 for i in range(len(df))]
    i = 0
    for id in df["ID"]:
        
        if(i%500 == 0):
            print("done", i, "on", len(df), end='\r')
        i+=1
        
        aut = df.loc[df["ID"] == id]
        df.loc[df.ID == id, "ending_year"] = get_ending_Y(pub_df.loc[pub_df["ID"] == id])
    return df
```

```python
# Print the distribution of ending publication years for each author
collaboration_df_with_ending_y = add_ending_y(collaboration_df, publications_df)
collaboration_df_with_ending_y.to_csv("myDATA/00-collaboration_df_with_ending_years.csv")  
collaboration_df_with_ending_y
```

```python
# print the ending year distribution for each author

collaboration_df_with_ending_y = pd.read_csv("myDATA/00-collaboration_df_with_ending_years.csv")
num_auths = []
for y in YEARS:
    num_auths.append(len(collaboration_df_with_ending_y.loc[collaboration_df_with_ending_y['ending_year']==int(y)]))

fig = plt.figure(figsize=(20, 5))
fig.suptitle('#_authors for ending year', fontsize=15)
plt.plot(YEARS,num_auths)
plt.xlabel('years')
plt.ylabel('#_authors')
fig.savefig('myDATA/end_year_distribution')
plt.show()
```
