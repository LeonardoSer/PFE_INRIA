```python
import itertools
import random
import json
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path 
import sys
import threading
import os
from IPython.display import clear_output
```

```python
## Find the starting collaboration year for each author
def get_ending_Y(aut):
    YEARS = [str(year) for year in range(1990,2019)]
    last_val = aut["2018"].values[0]
    for y in YEARS[::-1]:
        val = aut[y].values[0]
        if(val != last_val):
            return str(int(y)+1)
        
# add starting collaboration year to each row of the dataset
def add_ending_y(df):
    df["ending_year"] = [0 for i in range(len(df))]
    i = 0
    for id in df["ID"]:
        
        if(i%500 == 0):
            print("done", i, "on", len(df), end='\r')
        i+=1
        
        aut = df.loc[df["ID"] == id]
        df.loc[df.ID == id, "ending_year"] = get_ending_Y(aut)
    return df
```

```python
min_size = 0
max_size = 28
YEARS = [str(year) for year in range(1990,2019)]  

# Print the distribution of ending collaboration years for each active author and for each hole lenght
for size in range(min_size, max_size+1):
    print((' \033[1m' + "HOLE SIZE "+str(size) + '\033[0m ').center(120, '='))
    path = 'myDATA/05-filtered_by_hole_size/filtered_by_hole_size_'+str(size)+'.csv'
    
    if os.path.exists(path):

        df = pd.read_csv(path)
        df = add_ending_y(df)
        
        for y in YEARS:
            num_auths = len(df.loc[df['ending_year']==y])
            print("for year ", y, "there are ", num_auths, "authors")
```
